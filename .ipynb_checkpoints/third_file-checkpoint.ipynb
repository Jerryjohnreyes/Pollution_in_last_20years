```python
def fill_outliers_None(frame):
    columnas = frame.columns
    for columna in columnas:
        quartiles = frame[columna].quantile([0.25, 0.50, 0.75])
        iqr = quartiles[0.75] - quartiles[0.25]
        above = quartiles[0.75] + (1.5 * iqr)
        below = quartiles[0.25] - (1.5 * iqr)
        for item in frame[columna]:
            if (item != None):
                if ( (item < below) | (item > above) ):
                    frame[columna][item] = None
                    
        print(columna, ' : ', frame[columna].isnull().sum() )
    frame.dropna(axis=0, how='any')
    return frame
```

```python
dataWO = fill_outliers_None(numeric_data)
```

```python
def percentofNull(frame):
    nulos = frame.isnull().sum().sum()
    percentageNull = nulos / (frame.shape[0] * frame.shape[1])
    print('Percentage of Null values in the frame is ',percentageNull, "%")
```
